```python
vectorizer = TfidfVectorizer(stop_words = 'english', max_features = 2000)
# Only takes keywords whose frequency is in the top 2000 of the corpus
```

```python
X = vectorizer.fit_transform(df['matters'])
```

```python
X
```

```python
print(X[0])
# Only gives elements that are non zero
```

```python
movie2idx = pd.Series(df.index, index = df['title'])
```

```python
movie2idx
```

```python
idx = movie2idx['Avatar']
idx
```

```python
query = X[idx]
query = query.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
score = cosine_similarity(query, X)
```

```python
score
```

```python
score = score.flatten()
score
```

```python
plt.plot(score)
```

```python
plt.plot(score[(-score).argsort()])
```

```python
sorted_scores = (-score).argsort()
sorted_scores
```

```python
recommended_idx = sorted_scores[1:6]
recommended_idx
```

```python
for idx in recommended_idx:
  print(df['title'][idx])
```

## Create a Function for Recommendation

```python
def recommender(title):
  idx = movie2idx[title]
  if type(idx) == pd.Series:
    idx = idx.iloc[0]
  query = X[idx]
  scores = cosine_similarity(query, X)
  scores = scores.flatten()
  recommended_idx = (-scores).argsort()[1:6]
  return df['title'].iloc[recommended_idx]
```

```python
recommender('The Calling')
```

```python
recommender('Avatar')
```

```python
recommender('The Avengers')
```
